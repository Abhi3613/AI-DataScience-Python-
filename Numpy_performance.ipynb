```python
import time
import sys
import numpy as np
import matplotlib.pyplot as plt

# -- Create Datasets
N = 1_000_000
py_list = list(range(N))
np_array = np.arange(N)

# 2 -- Measure List Performance
start_time = time.time()
list_squared = [x **2 for x in py_list]
list_time = time.time() - start_time

# ---------------------------------
# 3. Here Measure List Performance
# ----------------------------------
start_time = time.time()
np_squared = np_array ** 2
numpy_time = time.time() - start_time

# 4.-- Speedup Factor
speedup = list_time / numpy_time

# 5.Here Memory Usage

# List memory (approx)
list_memory = sys.getsizeof(py_list) + sum(sys.getsizeof(x) for x in py_list)

# NumPy memory (exact)
numpy_memory = np_array.nbytes

# -----------------------------
# 6. Print Results
# -----------------------------
print("===== PERFORMANCE COMPARISON =====")
print(f"List Time: {list_time:.6f} seconds")
print(f"NumPy Time: {numpy_time:.6f} seconds")
print(f"Speedup (List / NumPy): {speedup:.2f}x")

print("\n===== MEMORY USAGE =====")
print(f"Python List Memory: {list_memory / (1024**2):.2f} MB")
print(f"NumPy Array Memory: {numpy_memory / (1024**2):.2f} MB")

# -----------------------------
# 7. Bonus: Bar Plot
# -----------------------------
labels = ['Python List', 'NumPy Array']
times = [list_time, numpy_time]

plt.figure()
plt.bar(labels, times)
plt.title("Execution Time Comparison")
plt.ylabel("Time (seconds)")
plt.xlabel("Data Structure")

plt.show()
```
